```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine-clustering.csv',header=None,usecols=[0,1,2])
df.columns = ['Class label','Alcohol','Malic Acid']
df
```

```python
print(df['Alcohol'].dtype)
```

```python
df['Alcohol'] = pd.to_numeric(df['Alcohol'], errors='coerce')
```

```python
df['Alcohol'] = df['Alcohol'].astype('category').cat.codes
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
df['Malic Acid'] = pd.to_numeric(df['Malic Acid'], errors='coerce')
sns.kdeplot(df['Malic Acid'])
```

```python
sns.scatterplot(x=df['Alcohol'], y=df['Malic Acid'], color='blue')  # Blue for Alcohol
sns.scatterplot(x=df['Alcohol'], y=df['Malic Acid'], color='red')   # Red for Malic Acid
```
